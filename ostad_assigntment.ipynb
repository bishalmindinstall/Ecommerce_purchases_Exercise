```python
# Import Pandas and Numpy Library,read the csv file

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Cust_Purch_FakeData.csv")
```

```python
#Max and Min ages of data set and find the mean of data set
max_age = df['age'].max()
min_age = df['age'].min()
mean_age = df['age'].mean()
```

```python
print("Maximum Age:", max_age)
```

```python
print("Minimum Age:", min_age)
```

```python
print("Mean Age:", round(mean_age,2))
```

```python
#Most common customers name
df['full_name'] = df['first'] + " " + ['last']
```

```python
top_3_names = df['full_name'].value_counts().head(3)
print(top_3_names)
```

```python
#two customers have same phn number
duplicate_phones = df[df.duplicated(subset='phone', keep = False)].sort_values(by='phone')
duplicate_phones[['first','last','phone']]
```

```python
# profession of structural Engineer
count_structural = df[df['profession'] == 'Structural Engineer']
print(count_structural)
```

```python
#male customer are structural engineer
male_customer = df[(df['profession'] == 'Structural Engineer') & (df['gender'] == 'Male')]
print(male_customer)
```

```python
#find out the female structural engineers from province alberta
female_struct_alberta = df[(df['profession'] == 'Structural Engineer') & (df['gender'] == 'Female') & (df['province'] == 'AB')]
print(female_struct_alberta)
```

```python
#max,min and average spending
max_spending = df['price(CAD)'].max()
min_spending = df['price(CAD)'].min()
average_spending = df['price(CAD)'].mean()
```

```python
print("Maximum Spending:", round(max_spending,2))
print("Minimum Spending:", round(min_spending,2))
print("Average Spending:", round(average_spending,2))
```

```python
#did not spent anything customer
no_spend_customer = df[df['price(CAD)'] == 0]
print(no_spend_customer)
```

```python
#Those who spent 100CAD or more
high_value_customer = df[df['price(CAD)'] >= 100]
print(high_value_customer)
```

```python
#email associated with credit card number
filtered = df[df['cc_no'] == '5020000000000230']
email_count = filtered['email'].count()
print("Total Emails:", email_count)
```

```python
df['cc_exp'].dtype
```

```python
df['cc_exp'] = pd.to_datetime(df['cc_exp'], errors='coerce')
```

```python
#cards are expiring in 2019
#sum()
exp_2019 = df[df['cc_exp'].dt.year == 2019]['cc_exp'].count()
print("Cards expiring in 2019:", exp_2019)
```

```python
#vis as their credit card
visa_count = (df['cc_type'] == 'Visa').sum()
print("Number of Visa users:", visa_count)
```

```python
#find the customer spent 100CAD using visa
customers_visa_100 = df[
    (df['price(CAD)'] >= 100) &
    (df['cc_type'] == 'Visa')
]

print(customers_visa_100)
```

```python
#two most common profession
top_two = df['profession'].value_counts().head(2)
print(top_two)
```

```python
#top 5 popular email
top5 = df['email'].value_counts().head(5)
print(top5)
```

```python
#email with am.edu
count_am_edu = df['email'].apply(lambda x: x.lower().endswith('am.edu')).sum()

print("Number of customers using am.edu:", count_am_edu)
```
